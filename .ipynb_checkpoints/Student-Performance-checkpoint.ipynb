Student Information

    school: Student’s school
        GP = Gabriel Pereira
        MS = Mousinho da Silveira
    sex: Student’s gender
        F = Female
        M = Male
    age: Student’s age (15–22 years)
    address: Home address type
        U = Urban
        R = Rural
    famsize: Family size
        LE3 = ≤3 members
        GT3 = >3 members
    Pstatus: Parent’s cohabitation status
        T = Together
        A = Apart
    guardian: Student’s guardian (mother, father, other)
    absences: Number of school absences (0–93) 

Family and Home Life

    Medu: Mother’s education
        0 = None
        1 = Primary
        2 = 5–9th grade
        3 = Secondary
        4 = Higher
    Fedu: Father’s education
        0 = None
        1 = Primary
        2 = 5–9th grade
        3 = Secondary
        4 = Higher
    Mjob: Mother’s job (teacher, health, services, at_home, other)
    Fjob: Father’s job (teacher, health, services, at_home, other)
    famsup: Family educational support (yes or no)
    famrel: Quality of family relationships (1 = very bad to 5 = excellent)
    internet: Internet access at home (yes or no) 

School and Extracurricular Activities

    reason: Reason for choosing school (home, reputation, course, other)
    traveltime: Travel time to school
        1 = <15 min
        2 = 15–30 min
        3 = 30–60 min
        4 = >1 hour
    studytime: Weekly study time
        1 = <2 hrs
        2 = 2–5 hrs
        3 = 5–10 hrs
        4 = >10 hrs
    failures: Number of past class failures (n if 1 ≤ n < 3, else 4)
    schoolsup: Extra educational support (yes or no)
    paid: Extra paid classes in the subject (yes or no)
    activities: Participation in extra-curricular activities (yes or no)
    nursery: Attended nursery school (yes or no)
    higher: Aspires to pursue higher education (yes or no) 

Social and Lifestyle

    romantic: In a romantic relationship (yes or no)
    freetime: Free time after school (1 = very low to 5 = very high)
    goout: Going out with friends (1 = very low to 5 = very high)
    Dalc: Workday alcohol consumption (1 = very low to 5 = very high)
    Walc: Weekend alcohol consumption (1 = very low to 5 = very high)
    health: Current health status (1 = very bad to 5 = very good) 

Academic Performance

    G1: First period grade (0–20)
    G2: Second period grade (0–20)
    G3: Final grade (0–20, target variable)

# Data Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("student-performance.csv", sep=";")

#Check id any value is missing
print(data.isna().values.any())
print(data.isnull().values.any())
print(data.duplicated().values.any())
#Remove whitespaces
data=data.map(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
#Checkout data features
print(data.columns)
```

```python
#Checkout the data
pd.set_option('display.max_columns', None)
(data.head())
```

Lets drop features we dont really encounter in an edtech app or which is probably unnecessary: 
- School, Sex, Age, Parent statuses, FamSize, reason, guardian, paid, romantic, Dalc, Walc

```python
data=data.drop(columns=["school","sex","age","address","famsize","Pstatus","Medu","Fedu","Mjob","Fjob","reason","guardian","romantic","paid","famrel","Dalc","Walc"])
data.head(8)
```

```python
#Lets convert the boolean data columns to numerical: schoolsup 	famsup 	activities 	nursery 	higher 	internet
for col in data.columns:
    if(col in ["schoolsup","famsup","activities","nursery","higher","internet"]):
        data[col]=pd.Series((0 if val=="no" else 1) for val in data[col])

data.head()
```

```python
#Lets create a new atRisk column based off G3. Let 12/60 (60%) be the percentage required to pass.
data["atRisk"]=pd.Series((1 if grade<12 else 0) for grade in data["G3"])
#Lets drop the now un-necessary G3 column
data=data.drop("G3",axis=1)
data.head()
```

All data is now converted to numerical, perfect for applying regression.
G1 and G2 will ofc be the highest contributing factor to at risk, but let's explore the relationship with other data too.

```python
sns.heatmap(data.drop(columns=["G1","G2"]).corr(), cmap="coolwarm")
```
